```python
filename = "GOOGL_price_data.xlsx"
wBook = xlrd.open_workbook(filename)
googl = wBook.sheet_by_name("GOOGL") # get the first (0th) sheet
```

```python
googl
```

```python
df = pd.read_excel(filename, sheet_name = 'GOOGL')
df
```

```python
dailyChanges = []
for i in range(1,googl.nrows): 
    openValues = googl.cell_value(i, 1) 
    closeValues = googl.cell_value(i,4)
    dailyChange = closeValues - openValues
    dailyChanges.append(dailyChange)
df["Daily Change"] = dailyChanges
df["Daily Return"] = df["Adj Close"].pct_change()
```
